##  Разрабатотка стратегии взаимодействия с пользователями на основе аналитических данных для сети фитнес-центров «Культурист-датасаентист»

## Описание проекта

<div style="border:solid blue 2px; padding: 20px">

**Сеть фитнес-центров «Культурист-датасаентист» разрабатывает стратегию взаимодействия с пользователями на основе аналитических данных.  
    Распространённая проблема фитнес-клубов и других сервисов — отток клиентов.   
    Чтобы бороться с оттоком, отдел по работе с клиентами «Культуриста-датасаентиста» перевёл в электронный вид множество анкет пользователей.  
    Ваша задача — провести анализ и подготовить план действий по удержанию клиентов.**

## Цель исследования

Подготовить план действий по удержанию клиентов на основе проведнного анализа данных

## Задачи проекта

научиться прогнозировать вероятность оттока (на уровне следующего месяца) для каждого клиента; 

сформировать типичные портреты пользователей: выделить несколько наиболее ярких групп и охарактеризовать их основные свойства;

проанализировать основные признаки, наиболее сильно влияющие на отток;

сформулировать основные выводы и разработать рекомендации по повышению качества работы с клиентами:  
1) выделить целевые группы клиентов;  
2) предложить меры по снижению оттока;  
3) определить другие особенности взаимодействия с клиентами.

**'Churn'** — факт оттока в текущем месяце;  

``Текущие поля в датасете:``  
``Данные пользователя за предыдущий до проверки факта оттока месяц:``    
**'gender'** — пол  
**'Near_Location'** — проживание или работа в районе, где находится фитнес-центр  
**'Partner'** — сотрудник компании-партнёра клуба (сотрудничество с компаниями, чьи сотрудники могут получать скидки на абонемент — в таком случае фитнес-центр хранит информацию о работодателе клиента)

**'Group_visits'** — факт посещения групповых занятий  
**'Avg_class_frequency_total'** — средняя частота посещений в неделю за все время с начала действия абонемента

**'Avg_additional_charges_total'** — суммарная выручка от других услуг фитнес-центра: кафе, спорт-товары, косметический и массажный салон

## Модель прогнозировния оттока клиентов

### Разбивка данных на обучающую и валидационную выборки

```python
#отделим из датасета признаки и целевую переменную 
X = df.drop(columns = ['churn'])
y = df['churn']
```

```python
#разделим данные на обучающую и валидационную выборку
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
```

### Обучение модели логистической регрессией

```python
#создадим объект класса StandardScaler для обучающей выборки
scaler = StandardScaler()
#обучаем
X_train_st = scaler.fit_transform(X_train)
#стандартизируем
X_test_st = scaler.transform(X_test)
```

```python
#логистическая регрессия
lr_model = LogisticRegression(random_state=0)
lr_model.fit(X_train_st, y_train)
lr_predictions = lr_model.predict(X_test_st)
lr_probabilities = lr_model.predict_proba(X_test_st)[:,1]
print('Метрики для модели логистической регрессии:')
print('accuracy_score: {}\nprecision_score: {}\nrecall_score:{}'.format(
    accuracy_score(y_test, lr_predictions),
    precision_score(y_test, lr_predictions),
    recall_score(y_test, lr_predictions)
))
```

### Обучение модели случайным лесом

```python
#случайный лес
rf_model = RandomForestClassifier(n_estimators = 100, random_state = 0) 
rf_model.fit(X_train_st, y_train)
rf_predictions = rf_model.predict(X_test_st)
rf_probabilities = rf_model.predict_proba(X_test_st)[:,1]
print('\nМетрики для модели случайного леса:')
print('accuracy_score: {}\nprecision_score: {}\nrecall_score:{}'.format(
    accuracy_score(y_test, rf_predictions),
    precision_score(y_test, rf_predictions),
    recall_score(y_test, rf_predictions)
))
```

У модели, обученной логистической регрессией показатели метрик accuracy (доля правильных ответов алгоритма), precision (точность) и recall (полнота) выше, следовательно данная модель лучше для прогнозирования.

2. Стимулирование участия в групповых занятиях. Для этого можно разработать специальные программы и акции, предлагающие возможность бесплатных или со скидкой посещений групповых занятий. Например, "Приведи друга и получи бесплатное посещение группового занятия".

Можно предложить скидки на продление абонементов, бесплатные тренировки или подарки для клиентов, которые решат продолжить свой абонемент на длительный срок. ## Маркетинговые действия

1. Проведение опросов и исследований среди клиентов, которые планируют покинуть фитнес-клуб, чтобы выяснить причины их оттока. На основе полученных данных можно разработать более эффективные стратегии для удержания клиентов.
2. Предлагать индивидуальные программы тренировок и питания, которые будут адаптированы под потребности и цели каждого клиента. Это поможет создать более персонализированное и привлекательное предложение для клиентов.
3. Запуск программ лояльности и бонусных программ. Например, предоставление скидок или бесплатных тренировок для клиентов, которые постоянно посещают фитнес-клуб или приглашают новых клиентов.
4. Организация различных мероприятий и специальных тренировок для клиентов, чтобы создать общую атмосферу командного духа и укрепить сообщество внутри фитнес-клуба.
5. Создание привлекательного веб-сайта и активное использование социальных сетей для привлечения новых клиентов и поддержания интереса у существующих. Предоставление актуальной информации о новых тренировках, событиях и предложениях поможет удержать клиентов и сделать фитнес-клуб более привлекательным.
6. Постоянный контакт с клиентами через email-рассылки, SMS-уведомления и другие коммуникационные каналы. Регулярное информирование о предстоящих акциях, скидках и новостях поможет поддерживать интерес клиентов и снизить вероятность их оттока.
7. Обеспечение высокого качества обслуживания и тренировок. Тренеры и персонал фитнес-клуба должны быть внимательными и предупредительными к клиентам, обеспечивать профессиональную помощь и сопровождение в процессе тренировок.
8. Поддержка и мотивация клиентов с помощью программы персонального тренировочного плана, целевых достижений и регулярной обратной связи.
9. Вовлечение клиентов в активности клуба, такие как фитнес-челленджи, тематические мероприятия и групповые занятия.
10. Постоянный мониторинг и анализ эффективности маркетинговых действий и реакции клиентов. Это позволит корректировать стратегии и подходы, чтобы эффективно снизить отток клиентов.
