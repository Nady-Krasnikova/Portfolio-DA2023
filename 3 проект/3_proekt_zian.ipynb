### Отработка пропущенных значений

ceiling_height - высота пололков - у 38% обявлений (9195/23699*100) нет данных о высоте потолков, здесь ничего не поделаешь, оставляем без изменений  

floors_total - всего этажей - у 0,3% объявлений (86/23699*100) нет данных о этажности дома, здесь ничего не поделаешь, оставляем без изменений

airports_nearest - расстояние до ближайшего аэропорта - у 23,4% объявлений (5542/23699*100) нет данных о расстоянии до ближайшего аэропорта, здесь ничего не поделаешь, оставляем без изменений  

cityCenters_nearest - расстояние до центра города - у 23,2% объявлений (5519/23699*100) нет данных о расстоянии до центра города, здесь ничего не поделаешь, оставляем без изменений  

parks_around3000 - количество парков в пределах 3 км - у 23,2% объявлений (5518/23699*100) нет данных о количестве парков в радицсе 3 км, здесь можем заменить пропущенные значения на '0'  

parks_nearest - расстояние до ближайшего парка - у 65,9% объявлений (15620/23699*100) нет данных о расстоянии до ближайшего парка, здесь ничего не поделаешь, оставляем без изменений  

ponds_around3000 - количесвто водоемов в пределах 3 км - у 23,2% объявлений (5518/23699*100) нет данных о количестве парков в радицсе 3 км, здесь можем заменить пропущенные значения на '0' 

ponds_nearest - расстояние до ближайшего водоема - у 61,5% объявлений (14589/23699*100) нет данных о расстоянии до ближайшего водоема, здесь ничего не поделаешь, оставляем без изменений  

days_exposition - длительность размещения объявления - у 13,4% объявлений (3181/23699*100) нет данных о длительности размещения объявления, эти пропущенные значения мы можем заменить на медиану по известным данным остальных объявлений  


```python
#удаляем строки с пропущенными значениями в столбце 'locality_name'
data.dropna(subset=['locality_name'], inplace=True)
```

```python
#заменяем на 0 кол-во балконов в пропущенных значениях
data['balcony'].fillna(0, inplace=True)
```

```python
#заменяем на 0 кол-во парков в радиусе 3 км
data['parks_around3000'].fillna(0, inplace=True)
```

```python
#заменяем на 0 кол-во прудов в радиусе 3 км
data['ponds_around3000'].fillna(0, inplace=True)
```

описываю столбцы, где тип данных неверный и на какой его нужно изменить:  

- last_price - тип данных указан как float (число с запятой), нужно изменить на int (натуральное число) - так удобнее работать со значениями 

- first_day_exposition - тип данных указан как object (строка), нужно изменить на date (дата) - так как в ячейке указывается именно дата размещения объявления  

- floors_total - тип данных указан как float (число с запятой), нужно изменить на int (натуральное число) - измеряется целыми числами  

- is_apartment - тип данных указан как object (строка), нужно изменить на bool (булевое значение)  

- balcony - тип данных указан как float (число с запятой), нужно изменить на int (натуральное число) - измеряется целыми числами  
- parks_around_3000 - тип данных указан как float (число с запятой), нужно изменить на int (натуральное число) - измеряется целыми числами  

- ponds_around_3000 - тип данных указан как float (число с запятой), нужно изменить на int (натуральное число) - измеряется целыми числами  

- days_exposition - тип данных указан как float (число с запятой), нужно изменить на int (натуральное число) - измеряется целыми числами  


```python
#теперь займемся изменением типов данных на нужный нам
data['last_price'] = data['last_price'].astype('int64')
```

```python
data['first_day_exposition'] = pd.to_datetime(data['first_day_exposition'],format='%Y-%m-%dT%H:%M:%S')
```

```python
data['floors_total'] =  data['floors_total'].astype('Int8')
```

```python
data['is_apartment'] = data['is_apartment'].astype('bool')
```

```python
data['balcony'] = data['balcony'].astype('int64')
```

```python
data['parks_around_3000'] = data['parks_around_3000'].astype('int64')
```

```python
data['ponds_around_3000'] = data['ponds_around_3000'].astype('int64')
```

### Уникальные значения

```python
#посмотрим на количество уникальные значения в столбце locality_name
data['locality_name'].unique().shape[0]
```

```python
#посмотрим на 364 значения
data['locality_name'].unique()
```

Какие есть варианты для стандартизации уникальных значений:
- привести все буквы в единый реестр (нижний привычнее для чтения)
- букву ё заменить на е
- проверим после этого наличие явных дубликатов

```python
#приводим к нижнему регистру
data['locality_name'] = data['locality_name'].str.lower()
```

```python
#заменяем ё на е
data['locality_name'] = data['locality_name'].str.replace('ё', 'е')
```

```python
#проверяем наличие явных дубликатов
data.duplicated().sum()
```

```python
data['locality_name'].unique()
```

```python
#заменим городской поселок на поселок городского типа
data['locality_name'] = data['locality_name'].str.replace('городской поселок','поселок городского типа')
```

```python
data['locality_name'].unique()
```

### Добавление в таблицу новых столбцов

```python
#добавим столбец со стоимостью 1 м в квадрате
data['cost_per_sqm'] = data['last_price'] / data['total_area']
```

```python
#день недели публикации объявления (0 — понедельник, 1 — вторник и так далее);
data['dayofweek'] = data['first_day_exposition'].dt.dayofweek
```

```python
#месяц публикации объявления
data['month'] = data['first_day_exposition'].dt.month
```

```python
#год публикации объявления
data['year'] = data['first_day_exposition'].dt.year
```

```python
#тип этажа квартиры (значения — «первый», «последний», «другой»)
data['floor_category'] = data.apply(
    lambda x: 'первый' if (x.floor == 1) 
    else ('последний' if (x.floor == x.floors_total) & isinstance(x.floors_total,int) 
    else 'другой'), axis=1
)

data['floor_category_digit'] = data.apply(
    lambda x: 0 if (x.floor == 1) 
    else (2 if (x.floor == x.floors_total) & isinstance(x.floors_total,int) 
    else 1), axis=1
)
```

```python
#расстояние до центра города в километрах (переведите из м в км и округлите до целых значений)
data['city_centers_nearest_km'] = round(data['city_centers_nearest'] / 1000)
```

Посчитаем среднюю цену одного квадратного метра в 10 населённых пунктах с наибольшим числом объявлений

```python
top_10 = data.pivot_table(index='locality_name', values=['last_price','total_area'] ,aggfunc=['sum','count'])
top_10.columns=['last_price_sum','total_area_sum','last_price_count','total_area_count']
top_10.pop('total_area_count')
top_10.sort_values(by=['last_price_count'],ascending=False, inplace=True)
top_10 = top_10.iloc[:10]
top_10['price_per_sq_m'] = top_10['last_price_sum']/top_10['total_area_sum']
top_10.sort_values(by=['price_per_sq_m'],ascending=True, inplace=True)
top_10['price_per_sq_m'].plot(kind='barh')
```
