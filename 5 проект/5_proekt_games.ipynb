## Описание проекта

<div style="border:solid blue 2px; padding: 20px">
    
**Я работаю в интернет-магазине «Стримчик», который продаёт по всему миру компьютерные игры. Из открытых источников доступны исторические данные о продажах игр, оценки пользователей и экспертов, жанры и платформы (например, Xbox или PlayStation). Мне нужно выявить определяющие успешность игры закономерности. Это позволит сделать ставку на потенциально популярный продукт и спланировать рекламные кампании.
Передо мной данные до 2016 года. Представлю, что сейчас декабрь 2016 г., и я планирую кампанию на 2017-й. Нужно отработать принцип работы с данными. 
В наборе данных попадается аббревиатура ESRB (Entertainment Software Rating Board) — это ассоциация, определяющая возрастной рейтинг компьютерных игр. ESRB оценивает игровой контент и присваивает ему подходящую возрастную категорию, например, «Для взрослых», «Для детей младшего возраста» или «Для подростков».**

## Предобработка данных

### Импорт всех необходимых библиотек для работы

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

###  Открытие файла с данными

```python
#присваиваем к переменной data информацию 
data = pd.read_csv('games.csv')
```

```python
#смотрим первые 10 строк
data.head(10)
```

### Изучение общей информации

```python
#смотрим общую информацию
data.info()
```

### Приводим названия столбцов к "змеиному" регистру

```python
#приводим названия столбцов к нижнему регистру через метод lower()
data.columns = data.columns.str.lower()
```

### Ищем пропущенные значения и их количество

```python
#смотрим на пропущенные значения - их кол-во - с сортировкой
data.isna().sum().sort_values()
```

### Удаление неинформатиных записей

```python
#удалим строки где нет имени - такое кол-во не повлияет на результат
data.dropna(subset=['name'], inplace=True)
```

```python
#удалим строки где нет жанра - такое кол-во не повлияет на результат
data.dropna(subset=['genre'], inplace=True)
```

```python
#строки, где нет года релиза заполним значением 0
data['year_of_release'].fillna(0, inplace=True)
```

Строки, где нет значений в столбцах critic_score, user_score, rating - занимают значительную часть от общего кол-ва данных удалить мы их не можем - это исказит первичную информацию заполнить их средней или медианой тоже будет неверно - можно сильно промахнуться оставим пока эти пропуски как есть

```python
#смотрим на пропущенные значения после обработки
data.isna().sum()
```

### Проверка типов данных

```python
#смотрим тип данных
data.info()
```

### Преобразование типов данных

```python
#преобразуем тип данный в столбце year_of_release с float64 на int64 - там только целые числа
data['year_of_release'] = data['year_of_release'].astype('Int64')
```

```python
#преобразуем тип данный в столбце critic_score с float64 на int64 - там только целые числа
data['critic_score'] = data['critic_score'].astype('Int64')
```

### Возможные причины пропусков

1) Ошибка при выгрузке в столбцах имя и год выпуска - игра обязательно имеет эти данные.
2) Пропуски в оценках критиков и пользователей, а так же в рейтинге может быть фактом - не все игры имеют рейтинг.

```python
#разбираем столбец user_score согласно инструкции
#смотрим уникальные значения
data['user_score'].unique()
```

#### Замена значения на NaN

```python
#значение tbd значит рейтинг не определен, мы заменим его на nan
data.loc[data['user_score'] == 'tbd', 'user_score'] = np.nan
```

#### Изменение типа данных

```python
#теперь заменим тип данных с object на float64
data['user_score'] = data['user_score'].astype('float')
```

### Суммарные продажи по регионам

### Проверка гипотез

#### Первая гипотеза

Нулевая гипотеза - средние пользовательские рейтинги платформ Xbox One и PC одинаковые. Альтернативная гипотеза - cредние пользовательские рейтинги платформ Xbox One и PC не одинаковые.

```python
#посчитаем средний пользовательский рейтинг платформы
XOne = actual_data[(actual_data['platform'] == 'XOne') &  (actual_data['user_score'] > 0)]['user_score']#.mean().round()
PC = actual_data[(actual_data['platform'] == 'PC') &  (actual_data['user_score'] > 0)]['user_score']#.mean().round()
print('Средний пользовательский рейтинг платформы Xbox One:', XOne.mean())
print('Средний пользовательский рейтинг платформы PC', PC.mean())
```

```python
#на первый взгляд рейтинги одинаковые, но мы должны проверить гиптезу через t-тест, что бы посмотреть 
#вероятность получения результата при условии, что нулевая гипотеза верна
#порог достоверности установим на уровне 1%

from scipy import stats as st

alpha = .01

results = st.ttest_ind(XOne, PC, equal_var=False)

print('p-значение:', results.pvalue)


if (results.pvalue < alpha):
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу")
```

Это значит что мы оснавляем нулевую гипотезу - Средние пользовательские рейтинги платформ Xbox One и PC одинаковые

#### Вторая гипотеза

Нулевая гипотеза - средние пользовательские рейтинги жанров Action и Sports одинаковые. Альтернативная гипотеза - Средние пользовательские рейтинги жанров Action и Sports разные.

```python
#посчитаем средний пользовательский рейтинг жанров
Action = actual_data[(actual_data['genre'] == 'Action') &  (actual_data['user_score'] > 0)]['user_score']#.mean().round()
Sports = actual_data[(actual_data['genre'] == 'Sports') &  (actual_data['user_score'] > 0)]['user_score']#.mean().round()
print('Средний пользовательский рейтинг жанра Action:', Action.mean())
print('Средний пользовательский рейтинг жанра Sports', Sports.mean())
```

```python
#на первый взгляд рейтинги одинаковые, но мы должны проверить гиптезу через t-тест, что бы посмотреть 
#вероятность получения результата при условии, что нулевая гипотеза верна
#порог достоверности установим на уровне 1
alpha = .01

results = st.ttest_ind(Action, Sports, equal_var=False)

print('p-значение:', results.pvalue)


if (results.pvalue < alpha):
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу")
```

Это значит что мы отвергаем нулевую гипотезу и принимаем альтернативную - средние пользовательские рейтинги жанров Action и Sports разные.
